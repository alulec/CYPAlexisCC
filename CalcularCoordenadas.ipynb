```python
# Se cargan las librerias

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# Se definen todas las funciones necesarias

def obtenerPunto3D():
    print("Ingrese el punto: ")
    punto = input()
    #punto = punto[1:-1]
    punto = punto.split(',')
    return [float(punto[0]), float(punto[1]), float(punto[2])]

def obtenerPunto2D():
    print("Ingrese el punto: ")
    punto = input()
    #punto = punto[1:-1]
    punto = punto.split(',')
    return [float(punto[0]), float(punto[1])]

def graficarFigura3D(listaX, listaY, listaZ, tipoFugura):
    fig = plt.figure()
    ax1 = fig.add_subplot(111,projection='3d')
    ax1.set_xlabel("X")
    ax1.set_ylabel("Y")
    ax1.set_zlabel("Z")
    
    if(tipoFugura == 0): # para piramide
        x = np.array([listaX[:-1]])
        y = np.array([listaY[:-1]])
        z = np.array([listaZ[:-1]])
        ax1.plot_wireframe(x, y, z)

        x = np.array([[listaX[0], listaX[-2]]])
        y = np.array([[listaY[0], listaY[-2]]])
        z = np.array([[listaZ[0], listaZ[-2]]])
        ax1.plot_wireframe(x, y, z)

        for i in range(len(listaX) -1):
            x = np.array([[listaX[i], listaX[-1]]])
            y = np.array([[listaY[i], listaY[-1]]])
            z = np.array([[listaZ[i], listaZ[-1]]])
            ax1.plot_wireframe(x, y, z)

    if(tipoFugura == 1): # para cajas
        mitad = int(len(listaX)/2)

        x = np.array([listaX[:-mitad]])
        y = np.array([listaY[:-mitad]])
        z = np.array([listaZ[:-mitad]])
        ax1.plot_wireframe(x, y, z)
        x = np.array([[listaX[0], listaX[-(mitad + 1)]]])
        y = np.array([[listaY[0], listaY[-(mitad + 1)]]])
        z = np.array([[listaZ[0], listaZ[-(mitad + 1)]]])
        ax1.plot_wireframe(x, y, z)

        x = np.array([listaX[mitad:]])
        y = np.array([listaY[mitad:]])
        z = np.array([listaZ[mitad:]])
        ax1.plot_wireframe(x, y, z)
        x = np.array([[listaX[mitad], listaX[-1]]])
        y = np.array([[listaY[mitad], listaY[-1]]])
        z = np.array([[listaZ[mitad], listaZ[-1]]])
        ax1.plot_wireframe(x, y, z)

        for i in range(mitad):
            x = np.array([[listaX[i], listaX[(i + mitad)]]])
            y = np.array([[listaY[i], listaY[(i + mitad)]]])
            z = np.array([[listaZ[i], listaZ[(i + mitad)]]])
            ax1.plot_wireframe(x, y, z)
    plt.show()

def graficarFigura2D(listaX, listaY):
    for i in range(len(listaX) -1):
        plt.plot([listaX[i], listaX[i+1]],[listaY[i], listaY[i+1]], "b-")
    plt.plot([listaX[0], listaX[len(listaX)-1]], [listaY[0], listaY[len(listaX)-1]], "b-")
    plt.show()

def escalar2d(listaX, listaY):
    factorX = float(input("Escalamiento en X: "))
    factorY = float(input("Escalamiento en Y: "))

    for i in range(len(listaX)):
        listaX[i] = listaX[i]*factorX;
        listaY[i] = listaY[i]*factorY;

def trasladar2d(listaX, listaY):
    movimientoX = float(input("Mover en X: "))
    movimientoY = float(input("Mover en Y: "))

    for i in range(len(listaX)):
        listaX[i] = listaX[i] + movimientoX;
        listaY[i] = listaY[i] + movimientoY;

def rotar2d(listaX, listaY):
    pivoteX = float(input("Pivote X: "))
    pivoteY = float(input("Pivote Y: "))
    grados = float(input("Grados: "))
    grados = (np.pi*grados)/180 #lo pasa a radianes

    for i in range(len(listaX)):
        listaX[i] = pivoteX + (listaX[i] - pivoteX)*np.cos(grados) - (listaY[i] - pivoteY)*np.sin(grados)
        listaY[i] = pivoteY + (listaX[i] - pivoteX)*np.sin(grados) + (listaY[i] - pivoteY)*np.cos(grados)

def escalar3d(listaX, listaY, listaZ):
    factorX = float(input("Escalamiento en X: "))
    factorY = float(input("Escalamiento en Y: "))
    factorZ = float(input("Escalamiento en Y: "))

    for i in range(len(listaX)):
        listaX[i] = listaX[i]*factorX;
        listaY[i] = listaY[i]*factorY;
        listaZ[i] = listaZ[i]*factorZ;

def trasladar3d(listaX, listaY, listaZ):
    movimientoX = float(input("Mover en X: "))
    movimientoY = float(input("Mover en Y: "))
    movimientoZ = float(input("Mover en Z: "))

    for i in range(len(listaX)):
        listaX[i] = listaX[i] + movimientoX;
        listaY[i] = listaY[i] + movimientoY;
        listaZ[i] = listaY[i] + movimientoZ;

def rotar3d(listaX, listaY, listaZ):
    print("Opciones: x, y, z")
    eje = input("Eje sobre el cual rotar: ")
    pivoteX = float(input("Pivote X: "))
    pivoteY = float(input("Pivote Y: "))
    pivoteZ = float(input("Pivote Z: "))
    grados = float(input("Grados: "))
    grados = (np.pi*grados)/180 #lo pasa a radianes

    if(eje == "z"):
        for i in range(len(listaX)):
            listaX[i] = pivoteX + (listaX[i] - pivoteX)*np.cos(grados) - (listaY[i] - pivoteY)*np.sin(grados)
            listaY[i] = pivoteY + (listaX[i] - pivoteX)*np.sin(grados) + (listaY[i] - pivoteY)*np.cos(grados)
    elif(eje == "x"):
        for i in range(len(listaX)):
            listaY[i] = pivoteY + (listaY[i] - pivoteY)*np.cos(grados) - (listaZ[i] - pivoteZ)*np.sin(grados)
            listaZ[i] = pivoteZ + (listaY[i] - pivoteY)*np.sin(grados) + (listaZ[i] - pivoteZ)*np.cos(grados)
    elif(eje == "y"):
        for i in range(len(listaX)):
            listaZ[i] = pivoteZ + (listaZ[i] - pivoteZ)*np.cos(grados) - (listaX[i] - pivoteX)*np.sin(grados)
            listaX[i] = pivoteX + (listaZ[i] - pivoteZ)*np.sin(grados) + (listaX[i] - pivoteX)*np.cos(grados)

```

```python
# Se obtiene informacion requerida para graficar

numeroDimensiones = int(input("¿Cuantas dimensiones son requeridas?: "))
numeroPuntos = int(input("¿Cuantos puntos tiene la figura?: "))
numeroTransformaciones = int(input("¿Cuantas transformaciones?: "))
print("")

if(numeroDimensiones == 2):
    coorX = []
    coorY = []
    for i in range(numeroPuntos):
        print(i + ":")
        punto = obtenerPunto2D()
        coorX.append(punto[0])
        coorY.append(punto[1])
    graficarFigura2D(coorX, coorY)

if(numeroDimensiones == 3):
    opcion = int(input("0-> Piramide  1-> Caja :"))
    coorX = []
    coorY = []
    coorZ = []
    for i in range(numeroPuntos):
        print(i, end=":")
        punto = obtenerPunto3D()
        coorX.append(punto[0])
        coorY.append(punto[1])
        coorZ.append(punto[2])

    graficarFigura3D(coorX, coorY, coorZ, opcion)


```

```python
if(numeroDimensiones == 2):
    for i in range(numeroTransformaciones):
        eleccion = input("Opciones: R-> rotar, T-> trasladar, E-> escalar")
        
        if(eleccion == "R"):
            rotar2d(coorX, coorY)
        elif(eleccion == "T"):
            trasladar2d(coorX, coorY)
        elif(eleccion == "E"):
            escalar2d(coorX, coorY)
    
    graficarFigura2D(coorX, coorY)


if(numeroDimensiones == 3):
    for i in range(numeroTransformaciones):
        eleccion = input("Opciones: R-> rotar, T-> trasladar, E-> escalar")
        
        if(eleccion == "R"):
            rotar3d(coorX, coorY, coorZ)
        elif(eleccion == "T"):
            trasladar3d(coorX, coorY, coorZ)
        elif(eleccion == "E"):
            escalar3d(coorX, coorY, coorZ)

    graficarFigura3D(coorX, coorY, coorZ, opcion)
```
